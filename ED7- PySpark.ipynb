```python
# IMPORTAR A SPARKSESSION
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
# confirgurar minha variável spark
spark = SparkSession.builder.master('local').appName('Intro-pyspark').config('spark.ui.port', '4050').getOrCreate()
```

```python
spark
```

```python
# dataframes no spark
schema = ['id', 'nome', 'cpf', 'celular']

dados = [
    (1, 'Oswaldo da Silva', '999.999.999-99', '(99)9.9999-9999'),
    (1, 'Maria dos Santos', '888.888.888-88', '(88)8.8888-8888'),
    (1, 'Priscila Medeiros', '777.777.777-77', '(77)7.7777-7777'),
    (1, 'Roberto Diniz', '666.666.666-66', '(66)6.6666-6666'),
]

df = spark.createDataFrame(dados, schema)
```

```python
df.show()
```

```python
df.printSchema()
```

```python
schema2 = 'id INT, nome STRING, cpf STRING, celular STRING, cidade STRING'

dados2 = [
    (1, 'Oswaldo da Silva', '999.999.999-99', '(99)9.9999-9999', 'Sobral'),
    (1, 'Maria dos Santos', '888.888.888-88', '(88)8.8888-8888', 'Fortaleza'),
    (1, 'Priscila Medeiros', '777.777.777-77', '(77)7.7777-7777', 'Rio de Janeiro'),
    (1, 'Roberto Diniz', '666.666.666-66', '(66)6.6666-6666', 'Sobral')
]

df2 = spark.createDataFrame(dados2, schema2)
```

```python
df2.printSchema()
```

```python
# verificar a quantidade de linhas
df2.count()
```

```python
# visualizar as colunas
df2.columns
```

Os dataframes são imutáveis.

```python
# 1º transformação nos dados - SELECT
df2.select('nome', 'cpf').show()
```

```python
df2.show()
```

```python
df4 = df2.select('nome', 'cpf')
```

```python
df4.show()
```

```python
# 1º transformação nos dados - SELECT (2ª forma)
df2.select(F.col('nome')).show() 
```

```python
# para criar um backup do df
df_backup = df2
```

```python
# recuperando df
df2 = df_backup
```

```python
df2.show()
```

```python
# aplicando um filtro no dataframe
df2.filter(F.col('cidade') == 'Sobral').show()
```

```python
df2.select(F.col('nome'), F.col('cpf'), F.col('cidade')).filter(F.col('cidade') == 'Fortaleza').show()
```

```python
# aplicando um filtro no dataframe (outra forma)
df2.filter(df2.cidade == 'Sobral').show()
```

```python
# aplicando um filtro no dataframe (outra forma)
df2.filter("cidade = 'Sobral'").show()
```

```python
 # aplicando um filtro no dataframe (duas cidades)
df2.filter("cidade = 'Sobral' or cidade = 'Fortaleza'").show()
```

```python
# aplicando um filtro no dataframe (duas cidades)
df2.filter((F.col('cidade') == 'Sobral') | (F.col('cidade') == 'Fortaleza')).show()
```

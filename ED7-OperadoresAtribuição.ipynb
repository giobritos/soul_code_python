```python
# atribuição simples
numero = 10
salario = 1200.00
altura = 1.88
nome = 'Giovana'
ativo = True 
```

```python
# atribuição com soma
valor = 5
# agora vamos adicionar + 3 à variável valor, que deve ficar com 8
valor += 3

print(valor)
```

```python
# atribuição com subtração
valor = 5
# agora vamos subtrair - 3 à variável valor, que deve ficar com 2
valor -= 3

print(valor)
```

```python
# atribuição com multiplicação
valor = 5
# agora vamos multiplicar * 3 à variável valor, que deve ficar com 15
valor *= 3

print(valor)
```

```python
# atribuição com divisão
valor = 10
# agora vamos dividir / 3 à variável valor, que deve ficar com 3.3333...
valor /= 3

print(valor)
```

```python
# atribuição com divisão inteira
valor = 10
# agora vamos dividir // 3 à variável valor, que deve ficar com 3
valor //= 3

print(valor)
```

```python
# atribuição com módulo
valor = 10
# agora vamos modular % 2 à variável valor, que deve ficar com 0
valor %= 2

print(valor)
```

```python
# atribuição com exponenciação
valor = 2
# agora vamos exponenciar ** 3 à variável valor, que deve ficar com 8
valor **= 3

print(valor)
```

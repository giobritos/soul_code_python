7) Um palíndromo é uma palavra ou frase que ao ser lida da esquerda pra direita, ou da direita pra esquerda retorna o mesmo resultado. 
Ex:  ANA, RADAR, socorram-me subi no ônibus em Marrocos...

Crie um algoritmo em python que leia uma palavra ou frase e diga se é ou não um palíndromo. Desconsidere os espaços em branco e maiúsculas e minúsculas. 

```python
!pip install unidecode
import unidecode 

frase = unidecode.unidecode(input("Digite uma palavra ou frase e descubra se ela é um palíndromo: "))

frase1 = frase.replace("-", "")

if frase1[::-1].replace(" ", "").lower() == frase1.replace(" ", "").lower():
  print('Essa palavra/frase é um palíndromo.')
else:
  print('Essa palavra/frase não é um palíndromo.')
```

8) Crie um algoritmo em python que leia um número inteiro positivo N e preencha uma lista com os N primeiros números da sequência abaixo. O primeiro número da sequência é o 2. Após percorra a lista e imprima os números na tela.

2      3        5        9        17        33        65     129       257

```python
num = int(input("Digite a quantidade de números que você quer ver da sequência: "))
while num <= 0:
  num = int(input("Número inválido. Digite um número postivo: "))

lista = []
q = 2
a1 = 1
anterior = 2

for i in range(num):
  if i == 0:
    lista.append(anterior)

  else:
    pg = a1 * (q**(i-1))

    seq = anterior + pg
    anterior = seq 

    lista.append(seq)    

for i in lista:
  print(i, end = ' ')
```

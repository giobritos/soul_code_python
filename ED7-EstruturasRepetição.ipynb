```python
numero = int(input("Digite um número positivo: "))

if numero <= 0:
  print(f"O número digitado {numero} não é positivo")
else:
  print(f"O número digitado {numero} é positivo")
```

WHILE

```python
numero = int(input("Digite um número positivo: "))

while numero <= 0:
  numero = int(input(f"O número {numero} não é positivo. Digite um número maior que zero: ")) 

print("Você digitou um número válido!")
```

```python
# faça um algoritmo que escreva na tela 1000 vezes a palavra Bom dia!
contador = 1

while contador <= 1000:
  print(f"{contador}. Bom dia!")

  contador += 1
```

```python
# faça um algoritmo para escrever na tela os números de 0 à 500
i = 0

while i <=500:
  print(i, end = " ")

  i += 1
```

```python
# faça um algoritmo para escrever na tela os números de 300 a 100 em ordem decrescente
i = 300

while i >= 100:
  print(i, end =" ")

  i -= 1
```

```python
# faça um algoritmo para escrever na tela os números múltiplos de 4 
# entre 300 a 100 em ordem decrescente

i = 300

while i >= 100:
  if i % 4 == 0:
    print(i, end =" ")

  i -= 1
```

FOR

```python
# SINTAXE for contador in range(valor_inicial_contador,valor_parada_contador,passo)

for i in range(1, 11, 1):
  print(i, end = ' ')
```

```python
# usando apenas um parâmetro no range

for i in range(10): # por padrão: valor_inicial = 0 e passo = 1
  print(i, end = ' ')
```

```python
# usando apenas 2 parâmetros

for i in range(1,10): # por padrão: passo = 1
  print(i, end = ' ')
```

```python
# ex. 1) crie um programa em Python que escreva na tela números de 250 a 120 em ordem decrescente.

for i in range(250,119,-1):
  print(i, end = ' ')
```

```python
# ex. 2) crie um programa em Python que escreva na tela números de 1000 a 1500 em ordem crescente.

for i in range(1000, 1501, 1):
  print(i, end = " ")
```

```python
# ex. 3) crie um programa em Python que escreva na tela números de 150 a 1000 em ordem crescente 
# que são múltiplos de 3.

for i in range(150, 1001, 3):
  print(i, end = " ")

# ou

for j in range (150,1001):
  if j % 3 == 0:
    print(j, end = ' ')
```

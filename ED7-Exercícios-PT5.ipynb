10) Escreva um algoritmo que leia um valor inicial A e imprima a seqüência de valores do cálculo de A! e o seu resultado. Ex: 5! = 5 X 4 X 3 X 2 X 1 = 120

```python
num = int(input("Escolha um número e calcule seu fatorial: "))

while num < 0: # não existe fatorial de número negativo
  num = int(input("Valor inválido! Escolha um número e calcule seu fatorial: "))

n = 1 # começamos o contador com 1 e não 0, pois ele se tornará útil ao processo da multiplicação e no fatorial não se multiplica valores por 0
fat = 1 # criamos uma variável para o fatorial para ir se atulizando conforme roda o loop

while n <= num:
  if num > 0:
    fat *= n # fatorial se multiplica com o n até o n ser o valor escolhido por num, esse valor vai substituindo o valor de fat
    n += 1
  if num == 0: # fatorial de 0 é igual a 1
    fat = 1

print(f"\nO fatorial do número {num} é igual a {fat}.") 
```

3) Faça um Programa que leia 4 notas de aluno e armazene-as dentro de uma lista. Em seguida mostre as notas e a média das notas na tela.

```python
notas = []
soma = 0

for i in range(4):
  nota = float(input(f'Digite a {i+1}ª nota: '))
  
  notas.append(nota)
  soma += nota
  media = soma / 4

print('\n')
print(f'Suas notas foram: {notas[0]}, {notas[1]}, {notas[2]}, {notas[3]}.')
print(f'A média das notas listadas é de: {media:.2f}')
```

4) Faça um Programa que leia 20 números inteiros e armazene-os numa lista. Em seguida, crie mais 2 listas, uma para armazenar os números pares e outra para armazenar os números impares. Ao final, Imprima as três listas.

```python
lista = []
lista_par = []
lista_impar = []

for i in range(20):
  num = int(input(f'Digite o {i+1}º número inteiro de sua preferência: '))
  while num <= 0:
    num = int(input(f'Número inválido. Digite o {i+1}º número inteiro de sua preferência: '))
  
  lista.append(num)

  if num % 2 == 0:
    lista_par.append(num)
  else:
    lista_impar.append(num)
  
print('\n')
print(f"Lista geral: {lista}")
print(f"Lista par: {lista_par}")
print(f"Lista ímpar: {lista_impar}")
```

5) Faça um Programa que peça as quatro notas de 10 alunos junto com o seu nome. Calcule e armazene numa lista a média de cada aluno, bem como em outra lista armazene o nome do aluno. Ao final imprima o número de alunos com média maior ou igual a 7.0 junto com o seu nome e a média.

```python
nomes = []
medias = []
n = 0

for i in range(4):
  aluno = input(f"\nDigite o nome do {i+1}º aluno: ")
  nomes.append(aluno)
  n1 = float(input("Digite a nota 1: "))
  while n1 < 0 or n1 > 10:
    n1 = float(input("Nota inválida! Digite novamente a nota 1 do aluno: "))
  n2 = float(input("Digite a nota 2: "))
  while n2 < 0 or n2 > 10:
    n2 = float(input("Nota inválida! Digite novamente a nota 2 do aluno: "))
  n3 = float(input("Digite a nota 3: "))
  while n3 < 0 or n3 > 10:
    n3 = float(input("Nota inválida! Digite novamente a nota 2 do aluno: "))
  n4 = float(input("Digite a nota 4: "))
  while n4 < 0 or n4 > 10:
    n4 = float(input("Nota inválida! Digite novamente a nota 2 do aluno: "))
       
  calc_media = (n1 + n2 + n3 + n4) / 4
  medias.append(calc_media)

  if calc_media >= 7:
    n += 1

print("\n")
print(f"Totalizando {n} aluno(s) com média maior ou igual a 7.00.")

for i in medias:
  if i >= 7:
    j = medias.index(i)
    print(f'O(A) aluno(a) {nomes[j]} teve média {medias[j]:.2f}.')
```

6) Faça um Programa que leia uma lista de 5 números inteiros. Após, percorra a lista e calcule a soma e a multiplicação dos números. Ao final, mostre os números lidos, a soma dos números e a multiplicação entre eles.

```python
lst = []
soma = 0
mult = 1

for i in range(5):
  num = int(input(f'Digite o {i+1}º número inteiro de sua preferência: '))
  
  lst.append(num)

for i in lst:
  soma += i
  mult *= i
  
print('\n')
print(f'Os números escolhidos são: {lst}')
print(f'A soma desses números é igual a: {soma}')
print(f'A multiplicação desses números é igual a: {mult}')
```

7) Faça um Programa que leia uma lista A com 10 números inteiros, calcule e mostre a soma dos quadrados dos elementos da lista.

```python
lst = []
soma = 0

for i in range(10):
  num = int(input(f'Digite o {i+1}º número inteiro de sua preferência: '))
  
  lst.append(num)

for i in lst:
  i = i ** 2
  soma += i
  
print('\n')
print(f'Os números escolhidos são: {lst}')
print(f'A soma dos quadrados desses números é igual a: {soma}')
```

8) Faça um programa que leia o nome, a idade e a altura de 10 alunos da SoulCode. Em seguida determine quantos alunos com mais de 13 anos possuem altura inferior à média de altura desses alunos.

```python
nomes = []
idades = []
alturas = []
soma = 0
j = 0

for i in range(10):
  aluno = input(f"\nDigite o nome do {i+1}º aluno: ")
  nomes.append(aluno)
  idade = int(input(f"Digite a idade do {i+1}º aluno: "))
  idades.append(idade)
  altura = float(input(f"Digite a altura do {i+1}º aluno: "))
  alturas.append(altura)

for i in alturas:       
  soma += i
  media = soma / 10

for i in idades:
  if i > 13:
    n = idades.index(i)
    if alturas[n] < media:
      j += 1

print("\n")
print(f"A média das alturas é: {media:.2f}")
print(f"{j} aluno(s) com mais de 13 anos possuem altura inferior à média.")
```

```python
# operador de igualdade
num1 = 10
num2 = 5
num3 = 10
num4 = 7
texto = "segunda"

print(num1 == num2)
print(num1 == num3)
print(num4 == 7)
print(texto == "Segunda")
print(num3 == 7 + 3)
```

```python
# operador diferente
num1 = 10
num2 = 5
num3 = 10
num4 = 7
texto = "segunda"

print(num1 != 10)
print(texto != "Segunda")
```

```python
# operador menor que (<)
num1 = 10
num2 = 5
num3 = 10
num4 = 7

print(num1 < num2)
print(num3 < num4)
print(num4 < num2)
print(num1 < 15)
print(num2 < 5)
```

```python
# operador menor ou igual que (<=)
num1 = 10
num2 = 5
num3 = 10
num4 = 7

print(num1 <= num2)
print(num3 <= num4)
print(num4 <= num2)
print(num1 <= 15)
print(num2 <= 5)
```

```python
# operador maior que (>)
num1 = 10
num2 = 5
num3 = 10
num4 = 7

print(num1 > num2)
print(num3 > num4)
print(num4 > num2)
print(num1 > 15)
print(num2 > 5)
```

```python
# operador maior ou igual que (>=)
num1 = 10
num2 = 5
num3 = 10
num4 = 7

print(num1 >= num2)
print(num3 >= num4)
print(num4 >= num2)
print(num1 >= 15)
print(num2 >= 5)
```

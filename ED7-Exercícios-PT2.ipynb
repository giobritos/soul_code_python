```python
'''
Faça um algoritmo que leia um número inteiro e diga se ele é maior,
menor ou igual a 10
'''
num = int(input("Digite um número inteiro de sua peferência: "))

if num > 10:
  print(num, "é maior que 10.")
if num < 10:
  print(num, "é menor que 10.")
if num == 10:
  print(num, "é igual a 10.")
```

```python
'''
Faça um programa que leia dois números inteiros e 
apresente o maior deles na tela. Considere que os números são diferentes.
'''
num1 = int(input("Digite um número inteiro de sua peferência: "))
num2 = int(input("Digite outro número inteiro de sua peferência: "))

if num1 > num2:
  print(num1, "é o maior número!")
else:
  print(num2, "é o maior número!")
```

```python
'''
Faça um programa que leia dois números float e 
calcule as 4 operações básicas entre eles:
'''
num1 = float(input("Digite um número de sua peferência: "))
num2 = float(input("Digite outro número de sua peferência: "))

soma = num1 + num2
print("O resultado da soma desses dois números é:", soma)

sub = num1 - num2
print("O resultado da subtração desses dois números é:", sub)

mult = num1 * num2
print("O resultado da multiplicação desses dois números é:", mult)

if num2 == 0:
  print("Erro! Não existe divisão por 0!")
else:
  div = num1 / num2
  print("O resultado da divisão desses dois números é:", div)
```

```python
'''
Faça um algoritmo que leia um número inteiro e diga se ele é par ou ímpar
'''
num = int(input("Digite um número inteiro e descubra se ele é par ou ímpar: "))

if num % 2 == 0:
  print("O número", num, "é par!")
else:
  print("O número", num, "é ímpar!")
```

MÉTODOS DE LISTA

```python
# adicionar itens em uma lista
carros.append('Astra') # adiciona um elemento ao final da lista
carros.insert(0, 'Palio') # adiciona um elemento em um índice qualquer

print(carros)
```

```python
compras = ['Arroz', 'Feijão', 'Óleo' ]
compras2 = ['Maminha', 'Patinho']

compras.extend(compras2) # adiciona o que tem em compras 2 em compras

print(compras)
```

```python
# remover itens de uma lista
compras.remove('Feijão') # remove pelo elemento
print(compras)

compras.pop(0) # remove pelo índice
print(compras)

del compras[2] # também remove pelo índice
# particularidade do del: se você não definir índice, ele exclui a lista 
print(compras)

compras.clear() # diferentemente do del que exclui a lista, o clear limpa apenas o conteúdo
print(compras)
```

```python
# iterar sobre a lista
print(compras)

# forma 1 - for in:
for x in compras:
  print(x, end = ' ')

print("\n")

# forma 2 - for range:
print(len(compras))

print("\n")

for i in range(len(compras)):
  print(compras[i], end = ' ')

print("\n")

# forma 3 - while:
i = 0
while i < len(compras):
  print(compras[i], end = ' ')
  i += 1
```

```python
# ordenação de lista
frutas = ['melancia', 'laranja', 'abacate', 'limao', 'banana']
print(frutas)

frutas.sort()
print(frutas)

frutas = ['melancia', 'Abacaxi', 'laranja', 'abacate', 'limao', 'banana', 'Kiwi']
print(frutas)

frutas.sort()
print(frutas)

frutas.sort(key = str.lower) # ignora as caixas altas
print(frutas)

numeros = [19, 4, 26, -14, 0, 27, 100, 1, 59, -7]
print(numeros)

numeros.sort()
print(numeros)

numeros.sort(reverse = True)
print(numeros)
```

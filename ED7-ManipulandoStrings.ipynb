```python
frase = "O rato roeu a roupa do Rei de Roma"

print('roupa' in frase)
print('ROUPA' in frase) #o python é case sensitive
print('roeu' not in frase)
print('Brasil' not in frase)

if 'roeu' in frase:
  print('A palavra roeu está dentro da frase')

if 'Roeu' in frase:
  print('A palavra roeu está dentro da frase') #não apresenta resultado, pois python é case sensitive
```

```python
# fatiamento de strings
fruta = 'Abacate'

# fruta[posicao_inicial:posicao_final:passo]
print(fruta[0:4])
print(fruta[2:6])
print(fruta[-1:-4:-1])

print(fruta[0:7:2])
print(fruta[6:-5:-1])

print(fruta[0:]) #entende que é para ir até o final 
print(fruta[:6]) #do início até a posição 5

print(fruta[::])
print(fruta[::-1]) #inverte a string

len(fruta)
```

```python
# alterar conteúdo dentro de uma string
frutas = "Morango, Kiwi, Laranja, Goiaba, Morango"

frutas2 = frutas.replace("Morango", "Limão")

print(frutas)
print(frutas2)
```

```python
# verificar quantidade de uma determinada palavra em uma string

frutas.count("Morango")
```

```python
# concatenando strings
fruta1 = 'Morango'
fruta2 = 'Laranja'

fruta3 = fruta1 + fruta2

print(fruta3)
```

```python
#split
frutas = "Abacaxi, Morango, Kiwi, Laranja, Goiaba, Morango"

lista_frutas = frutas.split(",")

print(lista_frutas)
```

```python
#index
frutas = "Abacaxi, Morango, Kiwi, Laranja, Goiaba, Morango"

print(frutas.index("Morango")) #retorna a primeira ocorrencia da palavra mencionada
```

```python
#find
frutas = "Abacaxi, Morango, Kiwi, Laranja, Goiaba, Morango"

print(frutas.find("Banana")) #se não exitir a palavra ela retorna -1
print(frutas.find("Morango")) #se tiver, retorna o index 
```

```python
for i in range(20):
  n = float(input(f"Digite o {i+1}º número: "))

  if i == 0:
    maior = n
    menor = n

  if n > maior:
    maior = n

  if n < menor:
    menor = n

print(f"""
O maior número digitado é {maior}
e o menor número digitado é {menor}!""")
```

```python
base = int(input("Digite um número para ser a base: "))
exp = int(input("Digite um número para ser o expoente: "))

potencia = 1

for i in range(exp):
  potencia *= base

print(f"{base} ^ {exp} = {potencia}")
```

```python
n = int(input("Digite um número postivo: "))
while n < 1:
  n = int(input("Número inválido. Digite um número postivo: "))

soma = 0

for i in range(1, n+1):
  soma += i

print(f"\nA soma dos número de 1 até {n} é = {soma}.")
```

```python
somaPar = 0
somaImpar = 0

for i in range(10):
  n = int(input(f"Digite o {i+1}º número: "))
  while n < 0:
    n = int(input("Número digitado deve ser positivo! Digite novamente: "))

  if n % 2 == 0:
    somaPar += n
  else:
    somaImpar += n

print(f"""
A soma dos número pares é igual a: {somaPar},
e a soma dos números ímpares é igual a: {somaImpar}.
""")
```

```python
n = int(input("Digite um número postivo maior que zero: "))
while n < 1:
  n = int(input("Número inválido. Digite um número postivo: "))

soma_divisor = 0

for i in range(1, n):
  if n % i == 0:
    soma_divisor += i 
      
if soma_divisor == n:
  print(f"\nO número {n} é perfeito!")
else: 
  print(f"\nO número {n} não é perfeito!")
```

```python
qtd_jogadores = int(input("Digite o número de jogadores do seu time: "))
while qtd_jogadores <= 0:
  qtd_jogadores = int(input("Quantidade inválida. Digite um número postivo: "))

soma_alturas = 0

for i in range(qtd_jogadores):
  altura = float(input(f"Digite a altura do {i+1}º jogador em metros: "))
  while altura < 0 or altura > 3:
    altura = float(input("Digite um valor de altura válido: "))

  soma_alturas += altura
  media_alturas = soma_alturas / qtd_jogadores

print(f"\nA média das alturas dos jogadores do time é: {media_alturas:.2f} metros")
```

```python
n = int(input("Digite um número postivo: "))
while n <= 0:
  n = int(input("Número inválido. Digite um número postivo: "))

qtd_divisores = 0

for i in range(1, n+1):
  if n % i == 0:
    qtd_divisores += 1

if qtd_divisores == 2:
  print(f"O número {n} é primo.")
else:
  print(f"O número {n} não é primo.")
```

```python
# declarar uma função - com parâmetro
def par_ou_impar(numero):
  if numero % 2 == 0:
    print(f"{numero} é par!")
  else:
    print(f"{numero} é ímpar!")

par_ou_impar(10)
par_ou_impar(15)

n = int(input("Digite um número para saber se ele é par ou ímpar: "))
par_ou_impar(n)
```

```python
"""
FUNÇÕES
- Ler um número 
- Verificar se ele é par ou ímpar
- Ler outro número 
- Verificar se ele é positivo ou negativo
- Ler outro número 
- Verificar se ele é multiplo de 7
- Ler outro número 
- Verificar se ele é positivo ou negativo
- Ler outro número 
- Verificar se ele é multiplo de 7
- Ler outro número 
- Verificar se ele é par ou ímpar
- Ler outro número 
- Verificar se ele é positivo ou negativo
"""
def par_ou_impar(numero):
  if numero % 2 == 0:
    print(f"{numero} é par!")
  else:
    print(f"{numero} é ímpar!")

def positivo_ou_negativo(numero):
  if numero > 0:
    print('Positivo')
  elif numero == 0:
    print('O número é 0.')
  else:
    print('Negativo')

def verifica_se_multiplo_de_7(numero):
  if numero % 7 == 0:
    print("É múltiplo de 7.")
  else:
    print("Não é múltiplo de 7.")

n = int(input("Digite um número: "))
par_ou_impar(n)

n = int(input("Digite um número: "))
positivo_ou_negativo(n)

n = int(input("Digite um número: "))
verifica_se_multiplo_de_7(n)

n = int(input("Digite um número: "))
positivo_ou_negativo(n)

n = int(input("Digite um número: "))
verifica_se_multiplo_de_7(n)

n = int(input("Digite um número: "))
par_ou_impar(n)

n = int(input("Digite um número: "))
positivo_ou_negativo(n)
```

```python
# função para somar 2 valores e mandar de volta (return) pra quem chamou a função 
def soma(num1, num2):
  return f'O resultado de {num1} + {num2} é {num1 + num2}'

numero = soma(2, 7)
print(numero)
```

```python
# crie uma função para saber se um nome digitado é pthon
def verifica_nome(nome):
  if nome.lower() == 'python':
    return f'O nome digitado foi {nome}!'
  else:
    return f'O nome digitado não é python!'

nome = input("Digite o nome de uma linguagem de programação: ")
print(verifica_nome(nome))
```

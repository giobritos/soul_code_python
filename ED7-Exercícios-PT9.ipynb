1) As maçãs custam R 0,30 cada se forem compradas menos do que uma dúzia,e R 0,25 se forem compradas pelo menos doze. Escreva um programa que leia o número de maçãs compradas, calcule e escreva o valor total da compra.

```python
def calculaValorTotal(qtd_maca):
  if qtd_maca < 12:
    return qtd_maca * 0.30
  if qtd_maca >= 12:
    return qtd_maca * 0.25

qtd = int(input('Digite aqui a quantidade de maçãs que você irá comprar: '))
while qtd < 0:
  qtd = int(input('Quantidade inválida. Digite aqui a quantidade de maçãs que você irá comprar: '))

retorno = calculaValorTotal(qtd)

print(f"O valor total da compra é de R${retorno}")
```

2) Escreva um  programa para ler 3 valores inteiros (considere que não serão lidos valores	iguais)	e escrevê-los em ordem crescente.

```python
def colocar_ordem(lst):
  lst.sort()
  return lst
  
lista = []

for i in range(3):
  lista.append(int(input(f'Digite o {i+1}º número inteiro de sua preferência: ')))
  
ordem = colocar_ordem(lista)

for i in ordem:
    print(i, end = ' ')
```

3) Escreva um programa para ler 20 valores inteiros e	escrever o maior deles.	 Considere que	 o usuário não	informará valores iguais.

```python
def descobreMaior(numero):
  n = float(input(f"Digite o 1º número: "))
  maior = n

  for i in range(numero-1):
    n = float(input(f"Digite o {i+2}º número: "))

    if n >= maior:
     maior = n

  return maior

retorno = descobreMaior(20)
print(f"O maior número digitado é {retorno}!")  
```

4) Faça um programa que peça ao usuário um número entre 12 e 20. Se a pessoa digitar um número diferente, mostrar a mensagem "entrada inválida" e solicitar o número novamente. Se digitar correto mostrar o número digitado.

```python
def valida_intervalo(num):
  while num < 12 or num > 20:
    print("Entrada inválida!")
    num = int(input("Digite um número entre 12 e 20: "))

  if num >= 12 or num <= 20:
    return num

numero = int(input("Digite um número entre 12 e 20: "))

print(f'O número digitado foi: {valida_intervalo(numero)}')
```

5) Crie um programa que leia um número do teclado até que encontre um número igual a zero. No final, mostre a soma dos números digitados

```python
def calculaValores(numero):

  soma = numero

  while True:
    numero = int(input('Digite um número de sua preferência: '))
    soma += numero
      
    if numero == 0:
      return soma
      break
      

n1 = int(input('Digite um número de sua preferência: '))

retorno = calculaValores(n1)

print(f'O resultado da soma dos números digitados foi: {retorno}')
```

6) Faça um programa que receba dois números inteiros e gere os números inteiros que estão no intervalo compreendido por eles.

```python
def mostraIntervalo(n1, n2):
  lst = []
  if n1 < n2:
    for i in range(n1, n2+1):
      lst.append(i)
    return lst

  if n1 == n2:
    return n1

  if n1 > n2: 
    for i in range(n1, n2-1, -1):
      lst.append(i)
    return lst
  
num1 = int(input("Digite um número inteiro: "))
num2 = int(input("Digite outro número inteiro: "))

print(f'Seu intervalo de [{num1},{num2}] contem os números: {mostraIntervalo(num1, num2)}')
```

7) Foi feita uma pesquisa entre os habitantes de uma região. Foram coletados os dados de idade, sexo (M/F) e salário. Faça um programa que calcule e mostre:

a) A média dos salários do grupo;
b) A maior e a menor idade do grupo;
c) A quantidade de mulheres na região;
d) A idade e o sexo da pessoa que possui o menor salário;

Finalize a entrada de dados ao ser digitada uma idade negativa.

```python
lstidades = []
lstsexo = []
lstsalario = []
soma_salario = 0

while True:
  idade = int(input('Digite a sua idade: (Coloque um valor negativo para sair do programa) '))

  if idade < 0:
    break

  lstidades.append(idade)
  maior_idade = idade
  menor_idade = idade

  while True:
     sexo = input('Digite seu sexo: M - Masculino ou F - Feminino: ')
     if sexo.lower() == 'm' or sexo.lower() == 'f':
       break
  lstsexo.append(sexo)

  salario = float(input('Digite o valor do seu salário: '))
  while salario < 0:
    salario = float(input('Valor inválido. Digite o valor do seu salário: '))
  lstsalario.append(salario)

  menor_salario = salario

  soma_salario += salario
  media_salario = soma_salario / len(lstsalario)

for i in range(len(lstidades)):  
  if lstidades[i] >= maior_idade:
    maior_idade = lstidades[i]

  if lstidades[i] <= menor_idade:
    menor_idade = lstidades[i]  

for i in range(len(lstsalario)):
  if lstsalario[i] <= menor_salario:
    menor_salario = lstsalario[i] 

idade_menorsal = lstidades[lstsalario.index(menor_salario)]
sexo_menorsal = lstsexo[lstsalario.index(menor_salario)]


print(f'A média de salários do grupo é de: {media_salario}')
print(f'A maior idade do grupo é {maior_idade} anos e a menor idade do grupo é {menor_idade} anos.')
print(f'A quantidade de mulheres no grupo é de {lstsexo.count("f")} mulheres.')
print(f'A pessoa que possui o menor salário tem {idade_menorsal} anos e é do sexo {sexo_menorsal}.')
```

8) Faça um programa que peça para n pessoas a sua idade, ao final o programa deverá verificar se a média de idade da turma varia entre 0 e 25,26 e 60 e maior que 60; e então, dizer se a turma é jovem, adulta ou idosa, conforme a média calculada

```python
def verificaMedia(idade):
  idades = [idade]

  while idade < 0:
    idade = int(input("Digite uma idade válida para começar: "))

  while True:
    idade = int(input("Digite a idade do aluno: (Digite um número negativo para parar o programa) "))
    if idade < 0:
      break
    
    idades.append(idade)

    soma = sum(idades)

  media = soma / len(idades)

  if media >= 0 and media <= 25:
    return "A turma é jovem."
  elif media >= 26 and media <= 60:
    return "A turma é adulta."
  else:
    return "A turma é idosa." 

idade = int(input("Digite a idade do aluno: "))

retorno = verificaMedia(idade)

print(retorno)
```

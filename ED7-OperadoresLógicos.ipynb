```python
# operador and
idade = 25
altura = 1.70

print(idade <= 30 and altura >= 1.65)
print(idade <= 20 and altura >= 1.65)
print(idade <= 30 and altura >= 1.75)
print(idade <= 20 and altura >= 1.75)
```

```python
# operador or
idade = 25
altura = 1.70

print(idade <= 30 or altura >= 1.65)
print(idade <= 20 or altura >= 1.65)
print(idade <= 30 or altura >= 1.75)
print(idade <= 20 or altura >= 1.75)
```

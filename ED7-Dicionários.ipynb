```python
# dicionários - chave:valor
# no dicionário, a chave pode ser string

# declarar um dicionário vazio
clientes = {}

# índice do dicionário
emails = {'sp':'antonio@gmail.com', 'rj':'juliana@hotmail.com', 'ce':'luana@outlook.com'}

print(emails['sp'])
```

```python
# crie um dicionário contendo os valores de um carro a seguir: Gol, G5, Volkswagen

carro = {'nome':'Gol', 'modelo':'G5', 'marca':'Volkswagen'}

print(carro['modelo'])
print(carro['marca'])

# dicionários dentro de uma lista
lista_carros = [
    {'nome':'Gol', 'modelo':'G5', 'marca':'Volkswagen'},
    {'nome':'Ônix', 'modelo':'LTZ', 'marca':'General Motors'},
    {'nome':'Corolla', 'modelo':'XEI', 'marca':'Toyota'},
    {'nome':'Corolla', 'modelo':'Altis', 'marca':'Toyota'}
]

print(lista_carros)
print(lista_carros[0])
print(lista_carros[1]['modelo'])
print(lista_carros[2]['marca'])
```

```python
# lista dentro de um dicionário
pessoas = {'nome': 'João', 'idade': 26, 'cursos': ['Python', 'Java', 'HTML'], 'altura': 1.78}

print(pessoas['nome'])
print(pessoas['idade'])
print(pessoas['cursos'])
print(pessoas['cursos'][1])

pessoas['cursos'].append('php')
print(pessoas['cursos'])
```

```python
# uma família é composta por 3 filhos - deve-se chamar por filho1, filho2, e filho3
# o retorno será nome, idade, ano_escola, cursos
familia = {'filho1':{'nome':'João', 'idade':13, 'ano_escola':"oitavo ano", 'cursos':['Python', 'Java']},
           'filho2':{'nome':'Mateus', 'idade':14, 'ano_escola':"nono ano", 'cursos':['Python', 'HTML']},
           'filho3':{'nome':'Joaquim', 'idade':16, 'ano_escola':"terceiro ano", 'cursos':['HTML', 'PHP']}
           }
           
# outra forma de escrever:
familia = {'filho1':{
               'nome':'João', 
               'idade':13, 
               'ano_escola':8, 
               'cursos':['ingles', 'espanhol']
               },
           
           'filho2':{
               'nome':'Mateus', 
               'idade':14, 
               'ano_escola':9, 
               'cursos':['ingles', 'frances']
               },
           
           'filho3':{
               'nome':'Joaquim', 
               'idade':16, 
               'ano_escola':12, 
               'cursos':['espanhol', 'italiano']
               }
           }

print(familia['filho1'])
print(familia['filho2']['nome'])
print(familia['filho3']['cursos'])
print(familia['filho3']['cursos'][1])
print('O nome do filho 3 é ' + familia['filho3']['nome'])
```

```python
for i in pessoas:
    print(pessoas[i])

for i in familia:
    print(familia[i])
```

```python
# uma família é composta por 3 filhos - deve-se chamar por filho1, filho2, e filho3
# o retorno será nome, idade, ano_escola, cursos
familia = {'filho1':{'nome':'João', 'idade':13, 'ano_escola':"oitavo ano", 'cursos':['Python', 'Java']},
           'filho2':{'nome':'Mateus', 'idade':14, 'ano_escola':"nono ano", 'cursos':['Python', 'HTML']},
           'filho3':{'nome':'Joaquim', 'idade':16, 'ano_escola':"terceiro ano", 'cursos':['HTML', 'PHP']}
           }
           
# outra forma de escrever:
familia = {'filho1':{
               'nome':'João', 
               'idade':13, 
               'ano_escola':8, 
               'cursos':['ingles', 'espanhol']
               },
           
           'filho2':{
               'nome':'Mateus', 
               'idade':14, 
               'ano_escola':9, 
               'cursos':['ingles', 'frances']
               },
           
           'filho3':{
               'nome':'Joaquim', 
               'idade':16, 
               'ano_escola':12, 
               'cursos':['espanhol', 'italiano']
               }
           }

print(familia['filho1'])
print(familia['filho2']['nome'])
print(familia['filho3']['cursos'])
print(familia['filho3']['cursos'][1])
print('O nome do filho 3 é ' + familia['filho3']['nome'])
```

```python
# método keys () - retorna todas as chaves de um dicionário
print(pessoas.keys())

print("\n")

for i in pessoas.keys():
    print(i)
```

```python
# método values() - retorna todos os valores de um dicionário
print(pessoas.values())

print("\n")

for i in pessoas.values():
    print(i)
```

```python
# método items() - retorna um conjunto de chave e valor de um dicionário
print(pessoas.items())

print("\n")

for i in pessoas.items():
    print(i)

print("\n")

for chave, valor in pessoas.items():
    print(f'{chave} = {valor}')
```

```python
# criar o mesmo esquema porém com N carros - dicionário
carros = {}

qtd_carros = int(input("Qual a quantidade de carros você irá adicionar? "))

for i in range(qtd_carros):
    carro = {}
    carro.update({
        'nome':input("Digite o nome no carro: "),
        'modelo':input("Digite o modelo do carro: "),
        'fabricante':input("Digite o fabricante do carro: "),
        'cor':input("Digite a cor do carro: "),
        'qtd_portas':int(input("Digite a quantidade de portas: "))
    })
    carros.update({f"carro {i+1}":carro})

print(carros)
print('\n')

for chave, valor in carros.items():
    print('\n')
    for x, y in valor.items():
        print(f'{x} = {y}')  
```

```python
# solução professor - lista

carros = []

while True:
    carros.append({
        'nome':input("Digite o nome no carro: "),
        'modelo':input("Digite o modelo do carro: "),
        'fabricante':input("Digite o fabricante do carro: "),
        'cor':input("Digite a cor do carro: "),
        'qtd_portas':int(input("Digite a quantidade de portas: "))
    })

    c = input('Digite s para sair ou outra tecla para continuar: ')
    if c.lower() == 's':
        break

for x in carros:
  print("\n")
  for k, v in x.items():
    print(f'{k} = {v}')  
```

```python
# solução professor - dicionário
carros = {}

while True:
  c = {}
  codigo = input("Digite qual o código do carro: ")
  c.update({
    'nome':input("Digite o nome no carro: "),
    'modelo':input("Digite o modelo do carro: "),
    'fabricante':input("Digite o fabricante do carro: "),
    'cor':input("Digite a cor do carro: "),
    'qtd_portas':int(input("Digite a quantidade de portas: "))
  })
  carros.update({codigo:c})

  flag= input('Digite s para sair ou outra tecla para continuar: ')
  if flag.lower() == 's':
    break

print(carros)
print('\n')

for k, v in carros.items():
  print(f'{k} = {v}')  
```

```python
# removendo item de um dicionário
pessoas = {'nome': 'João', 'idade': 26, 'cursos': ['Python', 'Java', 'HTML'], 'altura': 1.78}
print(pessoas)

# forma 1 - remove pela chava 
pessoas.pop('cursos')
print(pessoas)

# forma 2 - remove o último item
pessoas.popitem()
print(pessoas)

# forma 3 - del pessoas
del pessoas['idade'] # lembrar de passar argumento, pois se não deleta dicionário inteiro
print(pessoas)

# forma 4 - limpa todos os dados do dicionário sem excluí-lo
pessoas.clear()
print(pessoas)
```

```python
# cópias de um dicionário
pessoas = {'nome': 'João', 'idade': 26, 'cursos': ['Python', 'Java', 'HTML'], 'altura': 1.78}

pessoas_backup = pessoas.copy() # usa o copy para guardar o dicionário original
print(pessoas)
print(pessoas_backup)

pessoas.update({'cpf':'000.222.000-22'})
print(pessoas)
print(pessoas_backup)
```

```python
# operadores aritméticos

#soma
num1 = 10
num2 = 5
soma = num1 + 3

num1 = num1 + 3

soma = num1 + num2

print(soma)
print(num1)

#subtração
num1 = 10
num2 = 5

sub = num1 - num2

print(sub)

#multiplicação
num1 = 10
num2 = 5

mult = num1 * num2

print(mult)

#divisão
num1 = 10
num2 = 5

div = num1 / num2

print(div)

#divisão inteira
num1 = 10
num2 = 3

div_int = num1 // num2

print(div_int)

#resto da divisão
#ele também faz a divisão dos números, porém o resultado mostrado é o do resto
#muito usado para definir se números são pares ou ímpares ao serem divididos por 2
num1 = 10
num2 = 2

mod = num1 % num2

print(mod)

if mod == 0:
  print("Número é par")

#exponenciação
num1 = 2
num2 = 3

pot = num1 ** num2

print(pot)
```

```python
#FAÇA UM PROGRAMA QUE LEIA UM NOME E DOIS NÚMEROS
#E IMPRIMA NA TELA O NOME LIDO E A SOMA DOS DOIS NÚMEROS 
print("----------- PROGRAMA 1 -----------")
nome = input("Digite seu nome: ")
num1 = int(input("Digite um número inteiro: "))
num2 = int(input("Digite outro número inteiro: "))

soma = num1 + num2

print("Seu nome é: " + nome) # + concatena apenas string
print("A soma dos dois números escolhidos é: ", soma) # usar , para outros tipos de variável
```

```python
#FAÇA UM PROGRAMA QUE LEIA DOIS NÚMEROS 
#E EXIBA NA TELA O PRODUTO DELES
print("----------- PROGRAMA 2 -----------")
num1 = int(input("Digite um número inteiro: "))
num2 = int(input("Digite outro número inteiro: "))

mult = num1 * num2

print("O produto desses dois números é: ", mult)
```

```python
# if e else
geladeira = 'fechada'

if geladeira == 'aberta':
  print('Pode pegar a comida!')
else:
  print('A geladeira está fechada e não é possível pegar comida!')

if geladeira == 'aberta':
  print('Pode pegar a comida!')
if geladeira != 'aberta':
  print('A geladeira está fechada e não é possível pegar comida!')

if geladeira == 'aberta':
  print('Pode pegar a comida!')
if geladeira == 'fechada':
  print('A geladeira está fechada e não é possível pegar comida!')
```

```python
# if elif e else (encadeado)
num = int(input("Digite um número inteiro: "))

if num > 10:
  print(num, "é maior que 10.")
elif num < 10:
  print(num, "é menor que 10.")
else:
  print(num, "é igual a 10.")
```

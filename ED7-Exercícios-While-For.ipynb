1) Crie um programa em Python que escreva na tela números de 10 a n que são múltiplos de 4, onde o n é informado pelo  usuário e deve ser maior que 10. 

```python
n = int(input("Digite um número maior que 10: "))
while n <= 10:
  n = int(input("Digite um número válido maior que 10: "))

for i in range(10, n+1):
  if i % 4 == 0:
    print(i, end = ' ')
```

2) Em uma turma há 10 alunos. Cada aluno tem 2 notas. Um professor precisa calcular a média das duas notas de cada aluno. Crie um programa que resolva este problema.

```python
for i in range(10):
  aluno = input(f"\nDigite o nome do {i+1}º aluno: ")
  n1 = float(input("Digite a nota 1: "))
  n2 = float(input("Digite a nota 2: "))
  while n1 < 0 or n1 > 10:
    n1 = float(input("Nota inválida! Digite novamente a nota 1 do aluno: "))
  while n2 < 0 or n2 > 10:
    n2 = float(input("Nota inválida! Digite novamente a nota 2 do aluno: "))
     
  media = (n1 + n2) / 2

  print(f"{i+1}. A média do(a) aluno(a) {aluno} foi de: {media:.2f}")
```

3) Utilizando a estrutura de repetição for, faça um programa em Python que receba 10 números e conte quantos deles estão no intervalo [10,20] e quantos deles estão fora do intervalo, escrevendo estas informações.

```python
soma_i1 = 0
soma_i2 = 0

for i in range(10):
  num = float(input(f"Digite o {i+1}º número: "))
  if num >= 10 and num <= 20:
    soma_i1 += 1
  else:
    soma_i2 += 1

print(f"""\nA quantidade de números entre [10,20] é: {soma_i1},
e fora do intervalo [10,20] é: {soma_i2}.""")
```

4) Crie um programa em Python que leia N números (N digitado pelo usuário) e mostre na tela a soma de todos os números lidos.

```python
qtde = int(input("Digite quantos números serão digitados: "))
while qtde < 0:
  qtde = int(input("Quantidade inválida. Digite um número maior ou igual a zero: "))

soma = 0

for i in range(qtde):
  num = float(input(f"Digite o {i+1}º número: "))
  soma += num

print(f"\nA soma dos {qtde} números digitados é de: {soma}")
```

```python
# declarando uma tupla
frutas = ('banana', 'abacaxi', 'laranja')
```

```python
# tuplas são imutáveis
frutas[0] = 'abacate' 
```

```python
frutas[0] # acesso igual a listas
```

```python
# para mudar tuplas: converter para lista e depois voltar para tupla
frutas2 = list(frutas)
type(frutas2)
frutas2.append('limão')

frutas = tuple(frutas2)
print(frutas)
```

```python
for i in frutas:
  print(i)
```
